## Collision Severity and Cause

In the following cells we will use Spark RDD transformations to analyze the leading causes of injury and fatality listed in the historical collision dataset.

```python
def collision_severity_and_cause(row):
    #returns a severity rating for fatalities and injuries and listed causes
    #as a tuple (fatal [1,0],injury [1,0],cause1, cause2,.., cause5)
    output = []
    severity = [0,0]
    if float(row['collision_NUMBER OF PERSONS KILLED']) > 0: severity[0]=1
    if float(row['collision_NUMBER OF PERSONS INJURED']) > 0: severity[1]=1
    for i in range(1,6):
        try:
            cause = [row['collision_CONTRIBUTING FACTOR VEHICLE %s' %str(i)]]
            cause.append(tuple(severity))
            output.append(tuple(cause))
        except:
            pass
    return output
#add up total injuries and fatalities based on listed cause (filter out records with no listed cause)    
causes = (collisions
              .flatMap(collision_severity_and_cause)
              .reduceByKey(lambda v1,v2: (v1[0]+v2[0],v1[1]+v2[1]))
              .filter(lambda (key,val): key))
causes.cache()
causes.take(10)
```

## Collision Severity and Victim

In the following cells we will use Spark RDD transformations to analyze the leading type of victim in injury and fatal collisions in the historical collision dataset.

```python
def collision_severity_and_victim(row):
    #returns a severity rating for fatalities and injuries and listed causes
    #as a tuple (fatal [1,0],injury [1,0],cause1, cause2,.., cause5)
    timestamp = dt.datetime.strptime(row['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S')
    output = []
    output.append( ('%s_Motorist_Fatal' % timestamp.year,float(row['collision_NUMBER OF MOTORIST KILLED'])) )
    output.append( ('%s_Motorist_Injury' % timestamp.year,float(row['collision_NUMBER OF MOTORIST INJURED'])) )
    output.append( ('%s_Cyclist_Fatal' % timestamp.year, float(row['collision_NUMBER OF CYCLIST KILLED'])) )
    output.append( ('%s_Cyclist_Injury' % timestamp.year, float(row['collision_NUMBER OF CYCLIST INJURED'])) )
    output.append( ('%s_Pedestrian_Fatal' % timestamp.year, float(row['collision_NUMBER OF PEDESTRIANS KILLED'])) )
    output.append( ('%s_Pedestrian_Injury' % timestamp.year, float(row['collision_NUMBER OF PEDESTRIANS INJURED'])) )
    return output

def coll_dict(row):
    k,v = row
    
    return {
                'Year': int(k.split('_')[0]),
                'Victim':k.split('_')[1],
                'Severity':k.split('_')[2],
                'Total':v
            }

victims = (collisions
               .flatMap(collision_severity_and_victim)
               .reduceByKey(lambda x,y: x+y)
               .map(coll_dict)
               .filter(lambda d: d['Year'] in range(2013,2016)))
victims.cache()

fatal = pd.DataFrame(victims.filter(lambda d: d['Severity']=='Fatal').collect())
injury = pd.DataFrame(victims.filter(lambda d: d['Severity']=='Injury').collect())

sns.set(font_scale=1.3)
fig = plt.figure(1,figsize=(20,20))

ax = plt.subplot(2,1,2)
ax = sns.barplot(x="Year", y="Total", hue="Victim", data=fatal, palette=sns.color_palette("YlOrRd_r",6), ci=None, hue_order=['Motorist','Pedestrian','Cyclist'])
ax.set_ylabel("Count")
ax.set_title("Total Fatalities")

ax = plt.subplot(2,1,1)
ax = sns.barplot(x="Year", y="Total", hue="Victim", data=injury, palette=sns.color_palette("YlOrRd_r",6), ci=None, hue_order=['Motorist','Pedestrian','Cyclist'])
ax.set_ylabel("Count")
ax.set_title("Total Injuries")

#adjust the x labels so they don't run over eachother
labels=[item.get_text().replace(' ','\n') for item in ax.get_xticklabels()]
ax.set_xticklabels(labels)
```

## Section 2. Location and Cluster Analysis

In the following cells we will attempt to identify clusters in the collisions data by location and time.

First we will develop separate Pandas DataFrames for collisions resulting in injuries and fatalities. We will then project the latitude and longitude locations of the collisions to cartesian, or X-Y, coordinates so we can plot our collisions data and search for hotspots and clusters.

We will also use Matplotlib path elements to plot New York City Police Department precinct boundaries for context.

```python
def inj_victim(rec):
    #input: collisions RDD record
    #output: a list of copies of the record for each type of injury victim (pedestrian, cyclist, or motorist)
    output = []
    if float(rec['collision_NUMBER OF PEDESTRIANS INJURED'])>0: 
        rec['victim']='Pedestrian'
        output.append(rec)
    if float(rec['collision_NUMBER OF CYCLIST INJURED'])>0: 
        rec['victim']='Cyclist'
        output.append(rec)
    if float(rec['collision_NUMBER OF MOTORIST INJURED'])>0:
        rec['victim']='Motorist'
        output.append(rec)
    return output
        
def fatal_victim(rec):
    #input: collisions RDD record
    #output: a list of copies of the record for each type of fatality victim (pedestrian, cyclist, or motorist)
    output = []
    if float(rec['collision_NUMBER OF PEDESTRIANS KILLED'])>0: 
        rec['victim']='Pedestrian'
        output.append(rec)
    if float(rec['collision_NUMBER OF CYCLIST KILLED'])>0: 
        rec['victim']='Cyclist'
        output.append(rec)
    if float(rec['collision_NUMBER OF MOTORIST KILLED'])>0:
        rec['victim']='Motorist'
        output.append(rec)
    return output   

def project_lat_lng(rec,lat,lng,p):
    #input: RDD record of python dict type, a latitude field name and longitude field name, projection method
    #outout: record with added coordinate fields from lat/lng projected to X-Y for plotting
    x,y = p(float(rec[lng]),float(rec[lat]))
    rec['X'] = x
    rec['Y'] = y
    
    return rec

def es_geoshape_to_geojson(doc,geoshape_field):
    #input: Elasticsearch document with a geoshape field name
    #output: List of geojson features of the same shape type, with ES document fields as properties
    output = []
    from copy import deepcopy
    if doc[geoshape_field]['type'].lower() == 'point':
        shapes=[geojson.Point(doc[geoshape_field]['coordinates'])]
    if doc[geoshape_field]['type'].lower() == 'linestring':
        shapes=[geojson.LineString(doc[geoshape_field]['coordinates'])]
    if doc[geoshape_field]['type'].lower() == 'polygon':
        shapes=[geojson.Polygon(doc[geoshape_field]['coordinates'])]
    if doc[geoshape_field]['type'].lower() == 'multipolygon':
        shapes = []
        for poly in doc[geoshape_field]['coordinates']:
            shapes.append(geojson.Polygon(poly[0]))

    #print geojson.is_valid(shape)
    for shape in shapes:
        props = deepcopy(doc)
        props.pop(geoshape_field,None)
        output.append(geojson.Feature(geometry=shape,properties=props))
    return output

def geojson_path(geo_feature,p=None):
    #input: Geojson feature
    #output: matplotlib path coordinate list 
    if p:
        verts = [p(*c) for c in geojson.utils.coords(geo_feature)]
    else:
        verts = list(geojson.utils.coords(geo_feature))
    
    codes = [Path.MOVETO]
    for i in range(len(verts)-2): codes.append(Path.LINETO)
    codes.append(Path.CLOSEPOLY)   

    return Path(verts, codes)
```

```python
## Add X and Y coordinates to the collisions data using an appropriate map projection
p = Proj(init='epsg:2263') #NAD83 / New York Long Island (ftUS)
xy_recs = collisions.map(lambda row: project_lat_lng(row,'collision_LATITUDE','collision_LONGITUDE',p))
xy_recs.cache()
#xy_recs.take(1)
```

```python
#Load the police precinct outlines
#Spark-Hadoop Configuration for reading and writing to ES
es_read_conf = { 
        "es.resource" : 'police/precincts', 
        "es.nodes" : ES_hosts, 
        "es.net.http.auth.user" : ES_username, 
        "es.net.http.auth.pass" : ES_password 
    }


#reads the Elasticsearch collision data index into a Spark RDD of format (id, json document), let's go ahead and remove the
#uneccesary key and just keep the json document values.
precincts = sc.newAPIHadoopRDD(
        inputFormatClass="org.elasticsearch.hadoop.mr.EsInputFormat",
        keyClass="org.apache.hadoop.io.NullWritable", 
        valueClass="org.elasticsearch.hadoop.mr.LinkedMapWritable", 
        conf=es_read_conf).map(lambda (k,v): v)

#Convert the geojson format of the precincts to Matplotlib path elements for graphing
precincts.cache()
paths = precincts.flatMap(lambda rec: es_geoshape_to_geojson(rec,'geometry')).map(lambda feat: geojson_path(feat,p))
paths.cache()
```

```python
#create pandas dataframes of X,Y locations of injuries and fatalities
injury_xy = (xy_recs
             .filter(lambda rec: float(rec['collision_NUMBER OF PERSONS INJURED'])>0)
             .flatMap(inj_victim) #identify type of victims
             .map(lambda rec: (dt.datetime.strptime(rec['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S').year,rec['victim'],rec['X'],rec['Y'])))
             #.filter(lambda rec: rec[0] in range(2013,2016))) #return just records for 2013, 2014, and 2015

inj_df = pd.DataFrame(injury_xy.collect())
inj_df.columns = ['Year','Victim','x','y']
#we will filter for 2013, 2014 and 2015 because they are complete years in our dataset
inj_filtered = inj_df[inj_df['Year'].isin([2013,2014,2015])]

fatal_xy = (xy_recs
             .filter(lambda rec: float(rec['collision_NUMBER OF PERSONS KILLED'])>0)
             .flatMap(fatal_victim) #identify type of victims
             .map(lambda rec: (dt.datetime.strptime(rec['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S').year,rec['victim'],rec['X'],rec['Y'])))
             #.filter(lambda rec: rec[0] in range(2013,2016))) #return just records for 2013, 2014, and 2015

fatal_df = pd.DataFrame(fatal_xy.collect())
fatal_df.columns = ['Year','Victim','x','y']
#we will filter for 2013, 2014 and 2015 because they are complete years in our dataset
fatal_filtered = fatal_df[fatal_df['Year'].isin([2013,2014,2015])]
```

## Searching for Clusters

We will use an unsupervised machine learning algorithm called Density-based spatial clustering of applications with noise (DBSCAN) to try to find clusters of injuries and fataltities amongst pedestrians, cyclists, and motorists in the 2016 data.

DBSCAN defines core nodes and outliers based on distance density. With DBSCAN we can identify the number of clusters and their centroids. Further reading on DBSCAN can be found [here](https://en.wikipedia.org/wiki/DBSCAN).

```python
def rec_datetime_parse(rec):
    #input: collisions RDD record
    #output: tuple (datetime, x, y, fatalities, injuries)
    return (dt.datetime.strptime(rec['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S'),
            rec['X'],
            rec['Y'],
            float(rec['collision_NUMBER OF PERSONS KILLED']),
            float(rec['collision_NUMBER OF PERSONS INJURED']))
xyTimeRDD = xy_recs.map(rec_datetime_parse)
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

data = inj_filtered.as_matrix(columns=['x','y'])
db = DBSCAN(eps=100,metric='euclidean').fit(data)
labels = db.labels_
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

print n_clusters_

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

fig = plt.figure(figsize=(20,20))
ax = plt.subplot(111)
for path in paths.collect():
    patch = patches.PathPatch(path, edgecolor='black',facecolor='none', lw=2)
    ax.add_patch(patch)

    
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = data[class_member_mask & core_samples_mask]
    n = xy.size
    #print n
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=1)
    
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.set_xlabel('')
ax.set_ylabel('')
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

data = inj_filtered.as_matrix(columns=['x','y'])
print data.size
bandwidth = estimate_bandwidth(data, quantile=0.2, n_samples=data.size/2)

db = MeanShift(bandwidth=bandwidth,cluster_all=False).fit(data)
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

print n_clusters_
```

## Time of Day and Year

In the following cells we will analyze the occurrence of injury and fatal collisions over time.

```python
#get an RDD of injury and fatality counts by year, month, and hour
def month_hour_fatalities_injuries(row):
    #returns a tuple of month, hour, and fatality/injury count
    timestamp = dt.datetime.strptime(row['grid_fullDate'],'%Y-%m-%dT%H:%M:%S')
    return (
                (timestamp.month,timestamp.hour),
                (float(row['grid_totalFatalities']),float(row['grid_totalInjuries']),1)
            )

#add up total injuries and fatalities based on listed cause (filter out records with no listed cause)    
timeseries = (grid
              .map(month_hour_fatalities_injuries)
              .reduceByKey(lambda v1,v2: (v1[0]+v2[0],v1[1]+v2[1],v1[2]+v2[2]))
              .map(lambda (k,v): (k,(v[0]/v[2],v[1]/v[2])))) #normalize total counts
timeseries.cache()
timeseries.take(10)
```

## Section 3. Pedestrians

The following cells analyze the severity and cause of pedestrian injuries and fatalities.

```python
#get an RDD of injury and fatality counts by year, month, and hour
def ped_month_hour_fatalities_injuries(row):
    #returns a tuple of month, hour, and fatality/injury count
    timestamp = dt.datetime.strptime(row['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S')
    return (
                (timestamp.month,timestamp.hour),
                (float(row['collision_NUMBER OF PEDESTRIANS KILLED']),float(row['collision_NUMBER OF PEDESTRIANS INJURED']),1)
            )

#add up total injuries and fatalities based on listed cause (filter out records with no listed cause)    
timeseries = (collisions
              .map(ped_month_hour_fatalities_injuries)
              .reduceByKey(lambda v1,v2: (v1[0]+v2[0],v1[1]+v2[1],v1[2]+v2[2]))
              .map(lambda (k,v): (k,(v[0]/v[2],v[1]/v[2])))) #normalize total counts
timeseries.cache()
timeseries.take(10)

#pedestrianfatality rate by month and hour
fatalTimeseries = pd.DataFrame(timeseries.map(lambda (k,v): (k[0],k[1],v[0])).collect())
fatalTimeseries.columns = ['Month','Hour','Fatalities']
fatalTimeseries = fatalTimeseries.pivot(index="Month", columns="Hour", values="Fatalities")

#pedestrian injury rate by month and hour
injTimeseries = pd.DataFrame(timeseries.map(lambda (k,v): (k[0],k[1],v[1])).collect())
injTimeseries.columns = ['Month','Hour','Injuries']
injTimeseries = injTimeseries.pivot(index="Month", columns="Hour", values="Injuries")
```

```python
#create a heatmap of total collisions by Month and hour
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.heatmap(fatalTimeseries, cmap="YlOrRd",yticklabels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
ax.set_title("Incidence of Pedestrian Fatality by Month and Hour")
```

```python
#create a heatmap of total collisions by Month and hour
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.heatmap(injTimeseries, cmap="YlOrRd",yticklabels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
ax.set_title("Incidence of Pedestrian Injury by Month and Hour")
```

```python
def weekday_hour_collisions(row):
    #returns a tuple of weekday, hour, and fatality/injury count
    timestamp = dt.datetime.strptime(row['collision_DATETIME_C'],'%Y-%m-%dT%H:%M:%S')
    return (
                (timestamp.weekday(),timestamp.hour),
                (float(row['collision_NUMBER OF PEDESTRIANS KILLED']),float(row['collision_NUMBER OF PEDESTRIANS INJURED']),1)
            )

#add up total injuries and fatalities based on listed cause (filter out records with no listed cause)    
timeseries = (collisions
              .map(weekday_hour_collisions)
              .reduceByKey(lambda v1,v2: (v1[0]+v2[0],v1[1]+v2[1],v1[2]+v2[2]))
              .map(lambda (k,v): (k,(v[0]/v[2],v[1]/v[2])))) #normalize total counts
timeseries.cache()

fatalities = pd.DataFrame(timeseries.map(lambda (k,v): (k[0],k[1],v[0])).collect())
fatalities.columns = ['Weekday','Hour','Fatalities']
fatalities = fatalities.pivot(index="Weekday", columns="Hour", values="Fatalities")

injuries = pd.DataFrame(timeseries.map(lambda (k,v): (k[0],k[1],v[1])).collect())
injuries.columns = ['Weekday','Hour','Injuries']
injuries = injuries.pivot(index="Weekday", columns="Hour", values="Injuries")
```

```python
#create a heatmap of total collisions by Month and hour
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.heatmap(fatalities, cmap="YlOrRd",yticklabels=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
ax.set_title("Incidence of Fatality by Weekday and Hour")
```

```python
#create a heatmap of total collisions by Month and hour
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.heatmap(injuries, cmap="YlOrRd",yticklabels=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
ax.set_title("Incidence of Injury by Weekday and Hour")
```
